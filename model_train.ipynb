## Data Standardization

```python
data.drop(columns=['Snfc'], inplace=True)
```

```python
datax = data.drop(columns=['Snet'])
```

```python
features = datax.columns
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(datax)
dfx1=scaler.transform(datax)
dfx1 = pd.DataFrame(dfx1)
```

```python
dfx1.head()
```

```python
dfy1 = data['Snet']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dfx1,dfy1,test_size=0.2, random_state=42) 
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score
mse = root_mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
mse , r2
```

# DecisionTree

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(max_depth=5)
dt.fit(X_train, y_train)    
```

```python
from sklearn.metrics import root_mean_squared_error,r2_score
```

```python
predictions = dt.predict(X_test)
mse = root_mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
mse , r2
```

```python
dump(dt, 'decision_tree_simple.h5')  # Save the model
```

# RandomForest Regressor 

```python
from sklearn.ensemble import RandomForestRegressor
rf =RandomForestRegressor(max_depth=10)
rf.fit(X_train, y_train)
```

```python
predictions = rf.predict(X_test)
mse = root_mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
mse , r2
```

```python
dump(rf, 'random_forest_simple.h5')  # Save the model
```

## ANN


```python
from tensorflow import keras
import tensorflow as tf
```

```python
ANN = keras.Sequential([
          keras.layers.Flatten(input_shape=(14,)),  # input layer (1)
          keras.layers.Dense(32, activation='relu'),  # hidden layer (2)
          keras.layers.Dense(16, activation='relu'),  # hidden layer (3)
          keras.layers.Dense(8, activation='relu'),  # hidden layer (4)
          keras.layers.Dense(1, activation='linear') # output layer (5)
      ])
ANN.compile(optimizer='adam',loss=tf.keras.losses.MeanSquaredError())
keras.callbacks.EarlyStopping(monitor='loss',
                                      min_delta=0.1,
                                      patience=3,
                                      verbose=1, mode='auto')
```

```python
ANN.fit(X_train, y_train,epochs=200, batch_size=32, verbose=1)
```

```python
predictions = ANN.predict(X_test)
mse = root_mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
mse , r2
```

```python
ANN.save('ANN_model.h5')
```
