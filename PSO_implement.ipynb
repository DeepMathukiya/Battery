```python
import pandas as pd 
```

```python
data = pd.read_excel("dataset.xlsx", sheet_name="Run 1_ PV_battery_model")
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= data.drop(["time",'Snfc'] , axis=1)
```

```python
data.head()
```

```python
datax = data.drop(columns=['Snet'])
```

```python
features = datax.columns
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(datax)
dfx1=scaler.transform(datax)
dfx1 = pd.DataFrame(dfx1)
```

```python
dfy1 = data['Snet']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dfx1,dfy1,test_size=0.2, random_state=42) 
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(max_depth=  17, min_samples_split = 7)
dt.fit(X_train, y_train)  
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score
predictions = dt.predict(X_test)
mse = root_mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
mse , r2
```

```python
from joblib import dump
dump(dt, 'DecisionTreeRegressor_model.h5')
```

```python
from sklearn.ensemble import RandomForestRegressor
rf =RandomForestRegressor(n_estimators= 80, max_depth= 19)
rf.fit(X_train, y_train)
```

```python
predictions = rf.predict(X_test)
rmse = root_mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
rmse , r2
```

```python
dump(rf, 'RandomForestRegressor_model.h5')
```

```python
from xgboost import XGBRegressor
Xgb = XGBRegressor(n_estimators= 82, learning_rate= 0.1825958416173385)
Xgb.fit(X_train, y_train)
```

```python
predictions = Xgb.predict(X_test)
rmse = root_mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
rmse , r2
```

```python
dump(Xgb, 'XGBRegressor.h5')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(n_estimators= 84, learning_rate= 0.17822224171704676)
gb.fit(X_train, y_train)
```

```python
predictions = gb.predict(X_test)
rmse = root_mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
rmse , r2
```

```python
dump(gb, 'GradientBoostingRegressor_model.h5')
```
